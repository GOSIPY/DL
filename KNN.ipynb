```python
from sklearn import datasets,model_selection,metrics,neighbors
```

```python
date = datasets.load_digits()
```

```python
X,y = date.data,date.target
```

```python
model = neighbors.KNeighborsClassifier()
```

```python
model.fit(X,y)
```

```python
y_predict = model.predict(X)
```

```python
metrics.accuracy_score(y,y_predict)
```

```python
train_scores,test_scores = model_selection.validation_curve(model,X,y,param_name='n_neighbors',param_range=range(1,31),cv=5)
```

```python
%pylab inline
```

```python
fig,ax = plt.subplots(figsize=(15,5))
ax.plot(range(1,31),[mean(xxx) for xxx in test_scores],lw=2,ls='-',marker='*',label = 'test',c='g')
ax.plot(range(1,31),[mean(xxx) for xxx in train_scores],lw=2,ls='--',marker='o',label = 'train',c='k')
ax.legend(frameon=False, fontsize=16)
ax.grid()
```
