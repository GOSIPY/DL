```python
from sklearn import datasets,metrics,model_selection,neighbors
```

```python
data = datasets.load_iris()
```

```python
X,y = data.data,data.target
```

```python
model = neighbors.KNeighborsClassifier()
```

```python
model.fit(X,y)
```

```python
y_predict = model.predict(X)
```

```python
metrics.accuracy_score(y,y_predict)
```

```python
train_scores,test_scores = model_selection.validation_curve(model,X,y,param_name='n_neighbors',param_range=range(1,31),cv=5)
```
