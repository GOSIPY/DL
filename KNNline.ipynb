```python
from sklearn import model_selection,datasets,metrics,neighbors,linear_model
```

```python
data = datasets.load_iris()
```

```python
X,y = data.data,data.target
```

```python
b = []
for model,modelname in [[neighbors.KNeighborsClassifier(n_neighbors=3),'knn n=3'],
                        [neighbors.KNeighborsClassifier(n_neighbors=5),'knn n=5'],
                        [linear_model.LogisticRegressionCV(),'line_logistic']]:
    print(modelname)
    a = model_selection.cross_val_score(model,X,y,cv=5)
    print(a)
    b.append(a)
    
```

```python
fig,ax = plt.subplots(figsize=(15,10))
ax.plot(range(1,6),b[0],ls='-',lw=2,c='k',marker='*',label='knn n=3')
ax.plot(range(1,6),b[1],ls='--',lw=2,c='r',marker='*',label='knn n=5')
ax.plot(range(1,6),b[2],ls='--',lw=2,c='b',marker='*',label='line_logic')
ax.legend()
ax.grid()
```
